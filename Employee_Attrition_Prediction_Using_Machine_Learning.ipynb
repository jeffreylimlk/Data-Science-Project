```python

def convert_string_columns_to_dummies(df):
    """
    Identifies string columns in a DataFrame and converts them to dummy variables.

    Args:
        df: The input pandas DataFrame.

    Returns:
        The DataFrame with string columns converted to dummy variables.
    """

    # Get a list of string columns
    string_columns = df.select_dtypes(include=['object']).columns.tolist()

    # Create dummy variables for each string column
    for col in string_columns:
        dummies = pd.get_dummies(df[col], prefix=col, drop_first=True)  # Avoid dummy variable trap
        df = pd.concat([df, dummies], axis=1)
        df.drop(col, axis=1, inplace=True)  # Remove original string column

    return df

# Apply the function to your DataFrame
employee1_df = convert_string_columns_to_dummies(employee_df)
```

# TASK #7: TRAIN AND EVALUATE A LOGISTIC REGRESSION CLASSIFIER

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=42)
```

```python
X_train.shape
```

```python
X_test.shape
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score

LR = LogisticRegression()
LR.fit(X_train, y_train)
y_pred = LR.predict(X_test)
```

```python
y_pred
```

```python
from sklearn.metrics import confusion_matrix, classification_report

print("Accuracy {} %".format( 100 * accuracy_score(y_pred, y_test)))
```

```python
# Testing Set Performance
cm = confusion_matrix(y_pred, y_test)
sns.heatmap(cm, annot=True)
```

```python
print(classification_report(y_test, y_pred))
```

**MINI CHALLENGE #5:**
- **Train an XGboost algorithm instead of logistic regression**
- **Compare XGBoost model performance to Logistic Regression**

```python
from xgboost import XGBClassifier

model = XGBClassifier(use_label_encoder=False)
model.fit(X_train, y_train)
y_pred = model.predict(X_test)
```

```python
print("Accuracy {} %".format( 100 * accuracy_score(y_pred, y_test)))
```

```python
cm = confusion_matrix(y_pred, y_test)
sns.heatmap(cm, annot=True)
```

```python
print(classification_report(y_test, y_pred))
```
