```python
# Importing the libraries

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split

from sklearn.svm import SVC
from sklearn.model_selection import GridSearchCV
from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay, accuracy_score, f1_score, recall_score
import shap
import pyarrow
```

```python
# Reading the dataset
df = pd.read_parquet('final_dataset.parquet')

# Replacing NaN values with 0
df = df.fillna(0)
```

```python
df.head()
```

```python
# Visualizing where target variable is 1 i.e there is drought

plt.figure(figsize=(10, 6))
plt.scatter(df[df['target'] == 1]['date'], df[df['target'] == 1]['target'])
plt.title('Drought occurance')
plt.xlabel('Date')
plt.ylabel('Drought Condition as 1')
plt.show()
```

```python
# following are the features list

features = ['precip', 'max_temp', 'min_temp', 'gwl', 'daily_discharge_cfs','sm_pca', 'st_pca']
target = 'target'
```

```python
# Checking for multicollinearity between features

corr = df[features].corr()
corr.style.background_gradient(cmap='coolwarm')
```

We can see that there is correlation between max_temp and min_temp, but as per the domain knowledge, we know that these two features are important and we should keep them


```python
# Displaying the data before splitting

display(df[features].head())

display(df[[target]].head())
```

```python
# Splitting the dataset into the Training set and Test set

X_train, X_test, y_train, y_test = train_test_split(df[features], df[target], test_size = 0.2, random_state = 0)
```

```python
# No. of instance with 1 in target variable for training set
display(y_train.value_counts())

# No. of instance with 1 in target variable for test set
display(y_test.value_counts())
```

```python
# display the data after splitting

# Training set
display(pd.DataFrame(X_train).head())
display(pd.DataFrame(y_train).head())

# Test set
display(pd.DataFrame(X_test).head())
display(pd.DataFrame(y_test).head())
```

```python
# Scaling the features
# Feature scaling is performed to ensure that the model is not biased towards the features with higher magnitude

sc = StandardScaler()
X_train = sc.fit_transform(X_train)
```

```python
# Displaying the scaled training set

display(pd.DataFrame(X_train).head())
```

```python
# Model building
classifier_linear = SVC(kernel='linear')
#classifier_poly = SVC(kernel='poly')
#classifier_rbf = SVC(kernel='rbf')

# Fitting the model
classifier_linear.fit(X_train, y_train)
#classifier_poly.fit(X_train, y_train)
#classifier_rbf.fit(X_train, y_train)

# Predicting the Test set results
pred_linear = classifier_linear.predict(sc.transform(X_test))
#pred_poly = classifier_poly.predict(sc.transform(X_test))
#pred_rbf = classifier_rbf.predict(sc.transform(X_test))
```

```python
# Generating the confusion matrix for the linear kernel

cm = confusion_matrix(y_test, pred_linear)

disp = ConfusionMatrixDisplay(confusion_matrix=cm, display_labels=classifier_linear.classes_)
disp.plot()
disp.ax_.set_title('Confusion Matrix for Simple SVM with Linear Kernel')
```

```python
# Calculating the accuracy, f1 score and recall for the linear kernel 

accuracy_linear = accuracy_score(y_test, pred_linear)
f1_linear = f1_score(y_test, pred_linear)
recall_linear = recall_score(y_test, pred_linear)

# Adding the scores to a dataframe

scores = pd.DataFrame({'Kernel': ['Linear']})
scores['Accuracy'] = [accuracy_linear]
scores['F1 Score'] = [f1_linear]
scores['Recall'] = [recall_linear]

# Displaying the scores
display(scores)
```

### Model Optimization

#### Optimizing the model using GridSearchCV for Accuracy Metric

Optimizing a Support Vector Machine (SVM) model using `GridSearchCV` for the accuracy metric involves systematically exploring a range of **hyperparameters** to find the combination that yields the **highest accuracy on your dataset**. Heres a step-by-step breakdown of what this entails:

1) Define the `Hyperparameter Grid`:

For SVMs, common hyperparameters to tune include the *C* parameter (regularization), gamma (kernel coefficient for rbf, poly, and sigmoid kernels), and the kernel type itself (linear, poly, rbf, sigmoid).
You need to create a grid of values for these hyperparameters to test. For example:

**C**: **[0.1, 1, 10, 100]**

**gamma**: **[0.001, 0.01, 0.1, 1]**

**kernel**: **[linear, rbf]**

2) Use the `GridSearchCV` class from scikit-learn to perform an exhaustive search over the specified hyperparameter grid. You also specify the performance metric to optimize, which in this case is accuracy.

3) Fit the Model: Train the model using the fit method of `GridSearchCV`, which will evaluate all possible combinations of hyperparameters using cross-validation.

4) Select the Best Model: `GridSearchCV` will determine the best combination of hyperparameters based on the accuracy metric. The best model is the one that achieved the highest average accuracy across the cross-validation folds

5) Evaluate the Best Model: After identifying the best parameters, you can use this model to make predictions on your test data and evaluate its performance.

All in all, optimizing an SVM model using `GridSearchCV` for accuracy involves defining a range of hyperparameters to explore, setting up the grid search with accuracy as the scoring metric, fitting the model, and selecting the best-performing hyperparameter combination based on the accuracy achieved through cross-validation.

#### Optimizing the model using GridSearchCV for Accuracy Metric

```python
# Implementing SVM with GridSearchCV for hyperparameter tuning of linear, rbf and poly kernels

params_grid = [{'C': [0.1, 0.5, 1, 5, 10], 'kernel': ['linear']},
               {'C': [0.1, 0.5, 1, 5, 10], 'kernel': ['rbf']},
               {'C': [0.1, 0.5, 1, 5, 10], 'kernel': ['poly'], 'degree': [2, 3, 4]}]

classifier = SVC()
grid_search = GridSearchCV(estimator = classifier,
                           param_grid = params_grid,
                           scoring = 'accuracy', # accuracy is used as scoring metric
                           cv = 10,
                           n_jobs = -1)

grid_search.fit(X_train, y_train)

# Displaying the best parameters
best_accuracy = grid_search.best_score_
best_parameters = grid_search.best_params_
print('Best Accuracy:', best_accuracy)
print('Best Parameters:', best_parameters)
```
